```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

# Cargar el dataset
file_path = './forestfires.csv'
df = pd.read_csv(file_path)

# Cargar la imagen
image_path = './15-Table1-1.png'
img = Image.open(image_path)

# Crear la figura
plt.figure(figsize=(10, 8))

# Mostrar la imagen en el fondo con el eje Y invertido
plt.imshow(img, extent=[0, 9, 9, 0], aspect='auto')

# Calcular la densidad de los puntos en el plano X, Y con el eje Y invertido
sns.kdeplot(data=df, x='X', y='Y', cmap="Reds", fill=True, thresh=0, levels=100, alpha=0.6)

# Configurar el grfico para que coincida con la imagen, invirtiendo el eje Y
plt.xlim(0, 9)
plt.ylim(9, 0)  # Invertimos el eje Y
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Distribucin de Recuento de X e Y sobre la Imagen')

# Mostrar el grfico
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Cargar el dataset
file_path = './forestfires.csv'
df = pd.read_csv(file_path)

# Agrupar por coordenadas X, Y y sumar las reas quemadas
area_sum = df.groupby(['X', 'Y'])['area'].sum().reset_index()

# Cargar la imagen
image_path = './15-Table1-1.png'
img = Image.open(image_path)

# Crear la figura
plt.figure(figsize=(10, 8))

# Mostrar la imagen en el fondo
plt.imshow(img, extent=[0, 9, 9, 0], aspect='auto')  # Invertimos el eje Y

# Crear un scatter plot donde el tamao del punto refleje la suma del rea quemada
plt.scatter(area_sum['X'], area_sum['Y'], s=np.log1p(area_sum['area'])*10, c='red', alpha=0.6, edgecolor='black')

# Configurar el grfico para que coincida con la imagen
plt.xlim(0, 9)
plt.ylim(9, 0)  # Invertimos el eje Y para que el origen est en la parte superior izquierda
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Suma del rea Quemada sobre la Imagen')

# Mostrar el grfico
plt.show()

```
